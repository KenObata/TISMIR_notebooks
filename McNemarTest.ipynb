```python
BERT_split0_fine_tune = pd.read_csv(TransformerDIRECTORY +'McNemar_BERT_split0_fine_tune.csv')
BERT_split0_fine_tune = BERT_split0_fine_tune.drop(columns=['Unnamed: 0'])
BERT_split0_fine_tune
```
